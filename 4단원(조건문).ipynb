**조건문을 활용해 조건에 따라 다른 동작을 하는 코드를 작성할 수 있습니다. **

>--- 
>**💻 코드 미리보기:**
>
> **`if 조건:`**  
> if문입니다. 조건이 참이라면 내부의 코드가 실행됩니다.   
> **`elif 조건:`**  
> elif문입니다. if문의 조건이 거짓이고 elif문의 조건이 참인 경우에만 내부의 코드가 실행됩니다.   
> **`else:`**  
> 조건문의 가장 마지막 부분에 쓰이며, 앞의 조건들이 모두 거짓일 경우 내부의 코드들이 실행됩니다. 
> ---

## **조건문이란?**
* 조건에 따라 다르게 동작을 실행할 수 있도록 작성한 코드입니다.
* 조건문을 사용해, 조건이 참인 경우와 거짓인 경우를 나누어 각각 다른 코드가 실행되도록 명령할 수 있습니다.  



## **비교 연산자**
**📌 조건문과 비교 연산자**
* 조건문은 특정 조건이 참일 때 실행되는 코드를 지정하는 구문입니다.   
* 비교 연산자는 두 개의 값을 비교하여 참 또는 거짓을 반환합니다.   
* 따라서 비교연산자를 조건문의 조건으로 활용할 수 있습니다. 

**📌 비교 연산자 종류**
비교 연산자의 종류 | 설명 | 예시 | True of False
---|:---:|---|---
a>b | a가 b 초과일 때 참 | 17>5 | True
a>=b | a가 b 이상일 때 참 | 2>=4 | False
a==b | a와 b가 동일할 때 참 | 3==3 |True
a!=b | a와 b가 다른 경우 참 | 7!=7 | False
a<b | a가 b 미만일 때 참 | 3<4 |True
a<=b | a가 b 이하일 때 참 | 7<=9 | True



> **⚠️ 주의(WARNING):**
>
> * 비교 연산자 `==`과 대입 연산자 `=`을 헷갈리지 않게 주의해야 합니다.    
>
>> - a == b : a와 b가 동일할 때 참을 반환합니다.   
>> - a = b : 변수 a에 값 b를 대입해줍니다. (참고 : 이 코드 이전에 b가 선언되지 않았다면 오류가 발생합니다.)
>
> * !=, >=, <= 에서 "="는 반드시 오른쪽에 작성되어야 합니다. 


## **조건문 활용 방법**


### **조건문 활용 방법 1 : if문**


#### **기본 구조**  





**📌 작동 방식**
* if 옆에 작성된 조건이 참인 경우 if문 내부의 실행할 코드 1, 2가 실행됩니다.    
* 만약 조건이 거짓이라면, 조건문 내부의 코드가 실행되지 않고 다음 코드로 넘어가게 됩니다.

> **⚠️ 주의(WARNING):**
>
> * 조건 다음에는 반드시 콜론(:)을 붙여야 합니다.        
> * if문 내부 코드를 작성할 때에는 반드시 들여쓰기를 해주어야 합니다.   
> * 들여쓰기는 tab키를 한 번 누르거나 혹은 스페이스바를 4번 눌러 작성할 수 있습니다. 


#### **예제 1 : if문을 활용하여 n이 5인지 여부에 따라 다른 코드 실행하기**

```python
# 예제_1_if문을_활용하여_n이_5인지_여부에_따라_다른_코드_실행하기
n = int(input("숫자를 입력하세요"))
if n == 5:
    print("n은 5 입니다!")  # n이 5인 것이 참일 때 실행됩니다. 
                         
if n != 5:
    print("n은 5가 아닙니다!")  # n이 5가 아닌 것이 참일 때 실행됩니다. 
```

### **조건문 활용 방법 2 : if-else문**

#### **기본 구조** 




**📌 작동 방식**

* if 옆에 작성된 조건이 참이면 if문 내부 코드1, 2가 실행되고, 거짓이면 else문 내부 코드 3이 실행됩니다.  




> **⚠️ 주의(WARNING):**
>
>* if문은 조건을 반드시 작성해야 하지만, else문은 조건을 작성하지 않습니다. 
>* else 다음에도 콜론(:)을 반드시 작성해야 합니다. 

#### **예제 2 : if-else문을 활용하여 n이 5인지 여부에 따라 다른 코드 실행하기**

```python
# 예제_2_if_else문을_활용하여_n이_5인지_여부에_따라_다른_코드_실행하기

n = int(input("숫자를 입력하세요"))
if n == 5:
    print("n은 5 입니다!")  # n이 5라는 조건이 참일 때 실행됩니다. 
else:
    print("n은 5가 아닙니다!")  # n이 5라는 조건이 거짓일 때 실행됩니다. 
```

### **조건문 활용 방법 3 : if-elif-else문**




**📌 작동 방식**
* elif는 if문 조건 이외의 다른 조건에 대해 코드를 구성하고 싶을 때 사용합니다. 
* if 옆에 작성된 조건이 참이면 if문 내부 코드1, 2가 실행됩니다. 
* if문의 조건이 거짓이면 elif문 조건의 참, 거짓을 확인합니다. 
* elif문 조건이 참인 경우, elif문의 내부 코드 3, 4가실행됩니다.  
* if문과 elif문 조건이 모두 거짓인 경우 else이후의 코드 5가 실행됩니다. 

**📌 참고**
* elif문은 원하는 조건의 개수만큼 여러 개 추가될 수 있습니다.   
* else문은 생략해도 되지만, 예외적인 상황을 대비하기 위하여 작성해주는 것이 좋습니다.

#### **예제 3 : if-elif-else문을 활용하여 조건 3개에 대해 다른 코드 실행하기**

if-else-elif문을 활용하여 n이 5인 경우 "n은 5입니다!"를,   
5보다 큰 경우 "5보다 큽니다!"를, 5보다 작은경우 "5보다 작습니다!"를 출력해봅시다.

```python
# 예제_3_if_elif_else문을_활용하여_조건_3개에_대해_다른_코드_실행하기
n = int(input("숫자를 입력하세요"))
if n == 5:
    print("n은 5 입니다!")  # n==5 조건이 참일 때 실행됩니다. 
elif n > 5:
    print("n은 5보다 큽니다!")  # n==5 조건이 거짓이고, n>5 조건이 참일 때 실행됩니다. 
else:
    print("n은 5보다 작습니다!")  # n==5 조건이 거짓이고, n>5 조건도 거짓일 때 실행됩니다.
```
